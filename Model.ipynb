```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.utils import to_categorical

# ==========================
# Configuration Parameters
# ==========================
# Parameterize target ticker:
target_ticker = "AAPL"  # Change this to the desired ticker
target_pct_col = f'Pct_Change_{target_ticker}'  # e.g., "Pct_Change_AMZN"
target_col = "Target_Pct_Change"  # This will be our working column

folder_path = 'stock_data'
file_name = 'tech_data_with_sentiment.csv'
csv_path = os.path.join(folder_path, file_name)
          
window_size = 30                # Number of time steps in each sample
debug_mode = False              # If True, use a small subset (e.g., 20 samples) to debug/overfit
learning_rate = 0.001
batch_size = 64
max_epochs = 50
# Define thresholds for signal classification (these are examples)
threshold_buy = 0.4   # If the percentage change is greater than 0.2%, issue a Buy signal.
threshold_sell = -0.4  # If the percentage change is less than -0.2%, issue a Sell signal.

# ==========================
# 1. Load and Preprocess Data
# ==========================
df = pd.read_csv(csv_path, index_col=0, parse_dates=True)
print("Shape of df after reading CSV:", df.shape)
print("Columns in CSV:", df.columns.tolist())
print("Data sample:\n", df.head())

if df.empty:
    raise ValueError("The CSV file is empty. Please check the file content or path.")


# ==========================
# 2. Create Signal Column Based on Average Pct Change
# ==========================
# List of expected percentage change columns for each ticker:
if target_pct_col not in df.columns:
    raise KeyError(f"Expected column '{target_pct_col}' not found in DataFrame.")

# Create a new column for the target ticker's percentage change
df[target_col] = df[target_pct_col]

# Create signal based on the target ticker's percentage change
def classify_signal(pct_change):
    if pct_change > threshold_buy:
        return "Buy"
    elif pct_change < threshold_sell:
        return "Sell"
    else:
        return "Hold"

df['signal'] = df[target_col].apply(classify_signal)
print("Signal distribution:\n", df['signal'].value_counts())

# ==========================
# 3. Encode the Signal
# ==========================
# Map the signal strings to numeric labels
signal_mapping = {"Sell": 0, "Hold": 1, "Buy": 2}
df['signal_encoded'] = df['signal'].map(signal_mapping)

# Convert the encoded labels into one-hot vectors
# (This will transform values 0, 1, 2 into vectors like [1,0,0], [0,1,0], [0,0,1])
onehot_labels = to_categorical(df['signal_encoded'].values, num_classes=3)

# ==========================
# 4. Scale the Feature
# ==========================
# Identify numeric columns in your DataFrame excluding the target column
numeric_cols = df.select_dtypes(include=[np.number]).columns.drop(['signal_encoded', target_col])

# Scale all numeric features except the target column
scaler_features = MinMaxScaler(feature_range=(0, 1))
df[numeric_cols] = scaler_features.fit_transform(df[numeric_cols])

# Now scale the target column separately
scaler_target = MinMaxScaler(feature_range=(0, 1))
scaled_target = scaler_target.fit_transform(df[[target_col]])
df[f'{target_col}_Scaled'] = scaled_target

# Check scaled feature stats
print(df[[f'{target_col}_Scaled']].describe())

# ==========================
# 5. Create Sliding Window Dataset
# ==========================
def create_dataset(data, labels, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(labels[i+window_size])
    return np.array(X), np.array(y)

X_all, y_all = create_dataset(df[f'{target_col}_Scaled'].values, onehot_labels, window_size)
print(f"Dataset shapes -> X: {X_all.shape}, y: {y_all.shape}")

if debug_mode:
    n_debug = min(20, len(X_all))
    X_all, y_all = X_all[:n_debug], y_all[:n_debug]
    print("Debug mode enabled: Using a small subset for training.")

# ==========================
# 6. Split the Data
# ==========================
# For time series, split chronologically: training on all but the most recent sample.
X_train, y_train = X_all[:-1], y_all[:-1]
X_test, y_test = X_all[-1:], y_all[-1:]
print(f"Training shapes -> X: {X_train.shape}, y: {y_train.shape}")
print(f"Test shapes -> X: {X_test.shape}, y: {y_test.shape}")

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test  = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# ==========================
# 7. Build the Model
# ==========================
model = Sequential()
model.add(LSTM(64, activation='relu', return_sequences=True, input_shape=(window_size, 1)))
model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(LSTM(32, activation='relu', return_sequences=False))
model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

optimizer = RMSprop(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# ==========================
# 8. Callbacks and Training
# ==========================
early_stop = EarlyStopping(monitor='loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='loss', factor=0.5, patience=5, min_lr=1e-6)

history = model.fit(
    X_train, y_train,
    epochs=max_epochs,
    batch_size=batch_size,
    verbose=1,
    callbacks=[early_stop, reduce_lr]
)

# ==========================
# 9. Evaluation and Prediction
# ==========================
# Predict the probabilities on the test set.
prediction_prob = model.predict(X_test)[0]
# Determine the predicted class (0, 1, or 2)
predicted_class = np.argmax(prediction_prob)
signal_mapping_inv = {v: k for k, v in signal_mapping.items()}

print(f"\nPrediction probabilities for {df.index[-1].date()}: {prediction_prob}")
print("Predicted signal:", signal_mapping_inv[predicted_class])
print("Actual signal:", df['signal'].iloc[-1])


# ==========================
# 10. Plot Training Metrics
# ==========================
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Loss')
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Accuracy', color='orange')
plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
